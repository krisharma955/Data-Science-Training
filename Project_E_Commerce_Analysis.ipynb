```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('/content/gdrive/MyDrive/Online Sales Data.csv')

df = pd.DataFrame(data)
```

```python
# Best-Selling Products - According to Total Revenue

best_selling = (
    df.groupby('Product Name')['Total Revenue']
    .sum()
    .sort_values(ascending = False)
    .reset_index()
)
best_selling.head(10)
```

```python
# Best-Selling Products - According to Units Sold

best_selling_units = (
    df.groupby('Product Name')['Units Sold']
    .sum()
    .sort_values(ascending = False)
    .reset_index()
)
best_selling_units.head(10)
```

```python
# Monthy Analysis

df['Date'] = pd.to_datetime(df['Date'])

df['Month'] = df['Date'].dt.month_name()

monthy_sales = (
    df.groupby('Month')['Total Revenue']
    .sum()
    .sort_values(ascending = False)
    .reset_index()
)

monthy_sales
```

```python
# Revenue by Category

revenue_byCategory = (
    df.groupby('Product Category')['Total Revenue']
    .sum()
    .sort_values(ascending = False)
    .reset_index()
)

revenue_byCategory
```

```python
# Revenue By Region

region_analysis = (
    df.groupby('Region')['Total Revenue']
    .sum()
    .sort_values(ascending = False)
    .reset_index()
)

region_analysis
```

```python
# Revenue By Payment Method

payment_methodAnalysis = (
    df.groupby('Payment Method')['Total Revenue']
    .sum()
    .sort_values(ascending = False)
    .reset_index()
)

payment_methodAnalysis

print("\n")
```

```python
# Region & Product Category Analysis

region_product = pd.pivot_table(
    df,
    values = 'Total Revenue',
    index = 'Region',
    columns = 'Product Category',
    aggfunc = 'sum'
)

region_product
```

```python
# Monthly Revenue by Category

monthly_revenue = pd.pivot_table(
    df,
    values = 'Total Revenue',
    index = 'Month',
    columns = 'Product Category',
    aggfunc = 'sum'
)

monthly_revenue
```

```python
# Payment Method, Region wise Analysis

paymentMethod_region = pd.pivot_table(
    df,
    values = 'Total Revenue',
    index = 'Payment Method',
    columns = 'Region',
    aggfunc = 'sum'
)

paymentMethod_region
```

```python
# Units Sold Per Category

units_category = (
    df.groupby('Product Category')['Units Sold']
    .sum()
    .sort_values(ascending = False)
    .reset_index()
)

units_category
```

```python
# Purchasing Pattern - Monthly/Category/Units Sold

purchase_monthly = pd.pivot_table(
    df,
    values = 'Units Sold',
    index = 'Product Category',
    columns = 'Month',
    aggfunc = 'sum'
)

purchase_monthly
```

```python
# Purchasing Pattern - Category/Units Sold

category_units = (
    df.groupby('Product Category')['Units Sold']
    .sum()
    .reset_index()
)

category_units
```
