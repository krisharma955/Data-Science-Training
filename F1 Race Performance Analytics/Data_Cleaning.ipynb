# **Merging DataSets**

```python
master = results.copy()
```

```python
#Merging Drivers Data

master = master.merge(
    drivers[['driverId', 'forename', 'surname']],
    on='driverId',
    how='left'
)

master['driver_name'] = master['forename'] + " " + master['surname']

master.drop(columns=['forename', 'surname'], inplace=True)
```

```python
#Merging Constructors Data

master = master.merge(
    constructors[['constructorId', 'name']],
    on='constructorId',
    how='left'
)

master.rename(columns={'name': 'constructor'}, inplace=True)
```

```python
#Merging Races Data

master = master.merge(
    races[['raceId', 'year', 'round', 'circuitId', 'name', 'date']],
    on='raceId',
    how='left'
)

master.rename(columns={'name': 'race_name'}, inplace=True)
```

```python
#Merging Circuits Data

master = master.merge(
    circuits[['circuitId', 'name', 'location', 'country']],
    on='circuitId',
    how='left'
)

master.rename(columns={'name': 'circuit'}, inplace=True)
```

# **Feature Engineering** (Adding Columns to make Analysis easier)

```python
#Winner

master["winner"] = master["positionOrder"] == 1
```

```python
#Podium Finish

master["podium"] = master["positionOrder"] <= 3
```

```python
#Pole Position

master["pole_position"] = master["grid"] == 1
```

```python
#Positions Gained

master["positions_gained"] = master["grid"] - master["positionOrder"]
```

```python
#Finished Race

master["finished"] = master["statusId"] == 1
```

```python
#Fastest Lap

master["fastest_lap"] = master["fastestLap"].notna()
```

```python
#Speed

master["fastestLapSpeed"] = pd.to_numeric(
    master["fastestLapSpeed"],
    errors="coerce"
)
```

```python
master["date"] = pd.to_datetime(master["date"])
```

```python
master.isnull().sum()
```

The **fastestLapSpeed** column contains a large number of missing values because speed data is unavailable for many historical races and retired drivers. Since these missing values represent unavailable information rather than errors, they were retained.
