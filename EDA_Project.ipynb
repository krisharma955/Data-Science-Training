# **Visualization**

`**#1 What is the distribution of maximum daily temperatures (TMAX) during the study period?**`

````python
plt.figure(figsize=(10,6))

sns.histplot(
    data=df,
    x='TMAX',
    bins=25,
    color='#FF6B6B',
    edgecolor='black',
    kde=True
)

plt.title('Distribution of Maximum Temperature', fontsize=16, fontweight='bold')
plt.xlabel('Maximum Temperature (K)')
plt.ylabel('Frequency')

plt.show()
````

**Conclusion:** The maximum temperatures recorded in Rajasthan are primarily concentrated around 312–313 K, with the majority of observations falling between 310 K and 315 K. Extremely low and extremely high temperatures are relatively rare, suggesting that most days experience consistently high temperatures.

**`#2 How does solar radiation affect maximum temperature?`**

````python
plt.figure(figsize=(10,6))

sns.scatterplot(
    data=df,
    x='SRAD',
    y='TMAX',
    color='dodgerblue',
    alpha=0.5,
    s=40
)

plt.title('Solar Radiation vs Maximum Temperature',
          fontsize=16,
          fontweight='bold')

plt.xlabel('Solar Radiation')
plt.ylabel('Maximum Temperature (K)')

plt.show()
````

**Conclusion:** Maximum temperature tends to increase with solar radiation, indicating that incoming solar energy significantly influences atmospheric temperature.

**`#3 What percentage of days experienced heatwaves?`**

````python
heatwave_counts = df['HEATWAVE'].value_counts()

colors = ['#FF9999','#87CEEB']

plt.figure(figsize=(8,8))

plt.pie(
    heatwave_counts,
    labels=['No Heatwave','Heatwave'],
    autopct='%1.1f%%',
    colors=colors,
    startangle=90,
    shadow=True,
    explode=[0.02,0.08]
)

plt.title('Percentage of Heatwave Days',
          fontsize=16,
          fontweight='bold')

plt.show()
````

**Conclusion:** Heatwave events account for only about 5% of the total observations, whereas non-heatwave conditions make up the remaining 95%. This indicates that extreme heat events are relatively rare, although their effects can be severe and warrant close monitoring.

**`#4 Which month experiences the highest number of heatwave days?`**

````python
heatmap_data = df.pivot_table(
    values='HEATWAVE',
    index='YEAR',
    columns='MONTH',
    aggfunc='sum'
)

plt.figure(figsize=(6,8))

sns.heatmap(
    heatmap_data,
    annot=True,
    fmt='.0f'
)

plt.title('Heatwave Days by Year and Month')

plt.show()
````

**Conclusion:** Heatwave occurrences are concentrated during the summer months, particularly May and June. The frequency of heatwave days varies significantly across years, with 2024 experiencing the most severe heatwave activity. This highlights the seasonal nature of heatwaves and their changing intensity over time.

**`#5 Which districts experience the highest number of heatwave days?`**

````python
heatwave_district = df.groupby('DISTRICT')['HEATWAVE'].sum()

top_10_districts = heatwave_district.sort_values(ascending=False).head(10)

plt.figure(figsize=(12,7))

palette = (['#7FB3CC', '#F28E8E'] * 5)[:len(top_10_districts)]

sns.barplot(
    x=top_10_districts.index,
    y=top_10_districts.values,
    hue=top_10_districts.index,
    palette=palette,
    legend=False
)

plt.title('Top 10 Districts with Highest Heatwave Occurrence',
          fontsize=16,
          fontweight='bold')

plt.ylabel('Number of Heatwave Days')
plt.xlabel('District')
plt.xticks(rotation=45)

plt.show()
````

**Conclusion:** Bikaner and Kota experienced the highest number of heatwave days during the study period, indicating that these districts are particularly susceptible to extreme heat events. In contrast, districts such as Jaipur and Jodhpur recorded comparatively fewer heatwave days. This spatial variation highlights the unequal distribution of heatwave impacts across Rajasthan.
